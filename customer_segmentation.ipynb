```python
def find_outliers (df,variable:str): #esto le dice a la funcion que la variable es un string
    q1=df[variable].quantile(0.25) # calculamos el primer cuartil
    q3=df[variable].quantile(0.75) # calculamos el tercer cuartil
    iqr=q3-q1 #la diferencia entre cuartiles es el rango intercuartilico
    lw,uw=q1-(iqr*1.5),q3+(iqr*1.5) #los limites (bigotes/lineas verticales en el boxplot) se calculan multiplicando por 1.5
    #el 1.5 viene de la desviacion estandar, ya que contiene el 86% de los datos.
    #con una desviacion estandar es el 68% de los datos. con dos desv, estandar tenemos el 94% de los datos.
    df_outliers=df[(df[variable]<lw)|(df[variable]>uw)]
    return df_outliers.index #nos devuelve los indices de los valores con outliers
   
    
```
